```python
# K- Nearest Neighbour classfier
# The K - Distance based algorithm that drop a random datapoint based on default k it calculate the euclidean distance based on mejority of data points within nearest range would be our defined class.
#     E = /(x1-x2)^2 + (y1-y2)^2
```

```python
from sklearn.model_selection import train_test_split
from sklearn import datasets
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap
cmap = ListedColormap(['#FF0000', '#00FF00', '#0000FF'])

iris = datasets.load_iris()
X, y = iris.data, iris.target

X_train, X_val, y_train, y_val = train_test_split(X, y, stratify=y, test_size=0.2, random_state=43)

plt.figure(figsize=(10,6))
plt.scatter(X[:,3], y, c=y, cmap=cmap, edgecolors='k', s=20)
plt.show()
```

```python
from knn import KNN
clf = KNN(k=3)
clf.fit(X_train, y_train)
predictions = clf.predict(X_val)

accuracy = np.sum(predictions == y_val)/len(y_val)
print(f'Accuracy Score for KNN Model: {round(accuracy, 2)}')
```
