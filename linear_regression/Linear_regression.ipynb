- Relationship between Independend and Dependent wariable.

- Eq: Y = Wx + b  ( W - initial weight
                 b - bias
                 x - Data point)
- Cost Function = MSE - Mean Squared Error
            MSE = J(w,b) = 1/N sum(y(actual) - y(predicted)**2)

- J`(m,b) = [df/dw] _ [1/N -2xi(yi - (Wxi+b))]
          [df/db] - [1/N -2(yi - (Wxi+b))]
        ___This formula for Gradient which this
we can find the best minimum value for Gradient Descent
which is a iterative method where the initial weight and bias is assign and on gradient its moving showly to negative to find the global minima cost for model. 

with each iteration we have update new weights and bias
w(new) = w(old) - alpha(Learning Rate).dw 
b(new) = b(old) - alpha(learning rate).dw


```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn import datasets

X, y = datasets.make_regression(n_samples=100, n_features=1, noise=10, random_state=43)
X_train, X_val, y_train, y_val = train_test_split(X, y, test_size=0.2, random_state=431)

plt.figure(figsize=(10,6))
plt.scatter(X[:,0], y, color='b',marker='o', s=30)
plt.show()
```

```python
from linear_m import LinearRegression
regression = LinearRegression(lr=0.1)
regression.fit(X_train, y_train)
predicted = regression.predict(X_val)

def MSE(y_true, y_predicted):
    return np.mean((y_true-y_predicted)**2)

mse = MSE(y_val, predicted)
print(f'MSE Score for Linear Model: {mse}')
```

```python
y_pred_line = regression.predict(X)
cmap = plt.get_cmap('viridis')
plt.figure(figsize=(10,6))
ml1 = plt.scatter(X_train, y_train, color=cmap(0.8), s=10)
ml2 = plt.scatter(X_val, y_val, color = cmap(0.5), s=10)
plt.plot(X, y_pred_line, color='red', linewidth=2, label='Prediction')
plt.show()
```
